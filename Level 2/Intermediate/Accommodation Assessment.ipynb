# A Comprehensive Analysis on Cost of Accomodation among Young Adult

## Background
- In recent times, it is challenging for young adults between the age of 18 to 29 years, who are just starting their careers or who are getting into the phase of being independent. Knowing that many go to bigger cities to make ends meet or to search for a greener pasture. 

Task

- The purpose of this analysis is to know how much young African adults spend on accommodation in comparison to their incomes.

Dataset Description

- The data collection process took a week, and about 121 responses was gotten through google form. The data has 17 columns, Below are the columns name and their datatypes; 
 0. TimestamP   object 
 1. Age; int64  
 2. Gender object 
 3. Maximum Level of Education; object 
 4. Country; object 
 5. State/Province; object 
 6. Based on your answer above, how industrious is your State of residence; object 
 7. Currency; object 
 8. Monthly Salary in the currency you mentioned above; int64  
 9. Career Industry; object 
 10. Years of working experience; float64
 11. Do you foot your accommodation bills?; object 
 12. If you selected "Not completely", what fraction of the rent do you pay?; object 
 13. Accommodation Rented; object 
 14. Location of accommodation;  object 
 15. Cost of your accommodation per month; object 
 16. Do you think accommodation prices are generally high in your State of residence for young adults?; object 

Disclaimer
- This analysis is mainly for learning and may not be used in making crucial decisions because the sample of the population gotten cannot be generalised upon.

```python
def object_count_plot(df):
    """
    Description : Plot countplot for all categorical features present in the dataframe passed
    Argument : Dataframe
    """
    
    for var in data.columns:
        if data[var].dtype == 'object':
            print(data[var].value_counts())
            plt.figure(figsize=(10,5))
            g = sns.countplot(x=var,data=data)
            g.set_xticklabels(g.get_xticklabels(), rotation=90, ha="right")
            plt.tight_layout()
            plt.show()
```

```python
def numeric_distribution_plot(df):
    """
    Description : Gives distribution plot for all the numeric features in the dataframe passed
    Argument : Dataframe
    """
    for col in df.columns:
        if df[col].dtype != 'object':
            print(df[col].describe())
            plt.figure(figsize=(12,5))
            plt.title("Distribution of "+col)
            ax = sns.distplot(df[col].dropna())
            plt.tight_layout()
            plt.show()
```

# Data Exploration 

* This analysis is based solely on those who indicated that they foot their accommodation either fully or not completely 

```python
#select those who foot their accommodations.
data2 = data[(data['Do you foot your accommodation bills?'] == 'Yes') | (data['Do you foot your accommodation bills?'] == 'Not completely')]
data2.sample(5)
```

```python
#plot the categorical data of those who foot their accommodations.
object_count_plot(data2)
```

```python
#check for the average salary and cost of accommodation in relation to Age
salary=data2[['Age','Monthly Salary in the currency you mentioned above', 'Cost of your accommodation per month']]
SALARY =salary.groupby(['Age'],as_index=False).mean()
SALARY=SALARY.sort_values(['Age'], ascending=False)
SALARY
```

* The Age range of the data collected is between 18-29 years of age, the analysis here shows that an average 18-19 years majorly in Nigeria, does not have job nor foot their accommodations.

```python
fig, ax = plt.subplots(1,2, figsize=(15,5))
g = sns.barplot(x='Age', y='Monthly Salary in the currency you mentioned above', data=SALARY, ax=ax[0])
g.title.set_text('The Average monthly salary corresponding Age')

b =sns.barplot(x='Age', y='Cost of your accommodation per month', data=SALARY, ax=ax[1])
b.title.set_text('The Average Amount spent on accommodation in relation to Age')

plt.show()
```

* The average monthly salary is not in progression of Age, that can mean that Age is not a determining factor in compensation but competence. 
* The average amount spent on accommodation is relatively high betwwen 20 -21 years of age and relatively low in the late twenties.

```python
#check for the average salary and cost of accommodation in relation to Career/Industry
career_industry=data2[['Career Industry','Monthly Salary in the currency you mentioned above', 'Cost of your accommodation per month']]
Career_Industry =career_industry.groupby(['Career Industry'],as_index=False).mean()
Career_Industry=Career_Industry.sort_values(['Career Industry'], ascending=False)
Career_Industry
```

```python
fig, ax = plt.subplots(1,2, figsize=(15,8))
g = sns.barplot(x='Career Industry', y='Monthly Salary in the currency you mentioned above', data=Career_Industry, ax=ax[0])
g.set_xticklabels(g.get_xticklabels(), rotation=90, ha="right")
plt.tight_layout()
g.title.set_text('The Average monthly salary corresponding to Career Industry')

g = sns.barplot(x='Career Industry', y='Cost of your accommodation per month', data=Career_Industry, ax=ax[1])
g.set_xticklabels(g.get_xticklabels(), rotation=90, ha="right")
plt.tight_layout()
g.title.set_text('The Average Amount spent on accommodation in relation to Career Industry')

plt.show()
```

* In this Career/Industry analysis, the average spending on accommodation somewhat correspond to the the average monthly salary.

```python
#check for the average salary and cost of accommodation in relation to Level of Education
MaxLevelofEdu=data2[['Maximum Level of Education','Monthly Salary in the currency you mentioned above', 'Cost of your accommodation per month']]
MaxLevelOfEdu =MaxLevelofEdu.groupby(['Maximum Level of Education'],as_index=False).mean()
MaxLevelOfEdu=MaxLevelOfEdu.sort_values(['Maximum Level of Education'], ascending=False)
MaxLevelOfEdu
```

```python
fig, ax = plt.subplots(1,2, figsize=(12,6))
g = sns.barplot(x='Maximum Level of Education', y='Monthly Salary in the currency you mentioned above', data=MaxLevelOfEdu, ax=ax[0])
g.set_xticklabels(g.get_xticklabels(), rotation=90, ha="right")
plt.tight_layout()
g.title.set_text('Average Monthly Salary-Level of Education')

g = sns.barplot(x='Maximum Level of Education', y='Cost of your accommodation per month', data=MaxLevelOfEdu, ax=ax[1])
g.set_xticklabels(g.get_xticklabels(), rotation=90, ha="right")
plt.tight_layout()
g.title.set_text('Average Amount spent on Accommodation-Level of Education')

plt.show()
```

- The average monthly salary is high in those with Master's Degree, and low in Undergradutes. but the analysis shows that undergraduate spend more on accommodation. 

```python
##check for the average salary and cost of accommodation in relation to how industrious the state of residence
industrious=data2[['Based on your answer above, how industrious is your State of residence?','Monthly Salary in the currency you mentioned above', 'Cost of your accommodation per month']]
Industrious=industrious.groupby(['Based on your answer above, how industrious is your State of residence?'],as_index=False).mean()
Industrious=Industrious.sort_values(['Based on your answer above, how industrious is your State of residence?'], ascending=False)
Industrious
```

- The correlation between industriousness of state of residence and average monthly salary is postive, those who their place of residence are most industrious got highest average salary distribution and followed by just industrious and also, the spending on accommodation too is correlated
