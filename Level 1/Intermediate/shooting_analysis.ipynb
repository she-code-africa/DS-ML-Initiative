# Feature Engineering

```python
#date
import datetime
data['date'] = pd.to_datetime(data['date'])
```

```python
data['year'] = pd.DatetimeIndex(data['date']).year
data['month'] = pd.DatetimeIndex(data['date']).month
data['day'] = pd.DatetimeIndex(data['date']).day
```

```python
data.age.max(), data.age.min()
```

```python
#Age Group
bins = [0, 18, 45, 60, 100]
g_names = ['Teens', 'Adult', 'Old', 'Very_Old']

data['Age_Group'] = pd.cut(data['age'], bins, labels=g_names)
```

# Analysis by Age

```python
age_group=data[['Age_Group','id', 'manner_of_death']]
Age = age_group.groupby(['Age_Group', 'manner_of_death'],as_index=False).count()
Age=Age.sort_values(['Age_Group', 'manner_of_death'], ascending=False)
Age
```

```python
grouped_pivot = Age.pivot(index='Age_Group',columns='manner_of_death')
grouped_pivot.head()
```

```python
sns.barplot(y='id', x='Age_Group', hue='manner_of_death', data=age_group)
```

# Analysis by Race

```python
Race=data[['race','id', 'manner_of_death']]
RACE= Race.groupby(['race', 'manner_of_death'],as_index=False).count()
RACE=RACE.sort_values(['race', 'manner_of_death'], ascending=False)
RACE
```

```python
pivot_race = RACE.pivot(index='race',columns='manner_of_death')
pivot_race.head()
```

```python
colors_list =['#5bc0de', '#d9534f']
ax = pivot_race.plot(kind='bar', figsize=(12,4), width=0.8, color=colors_list, edgecolor=None)

for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy()
    ax.annotate(f'{height}', (x +width/2, y + height*1.02), ha='center')
```

# Analysis by Arms_category

```python
Arms_category=data[['arms_category','id', 'manner_of_death']]
Arms_Category= Arms_category.groupby(['arms_category', 'manner_of_death'],as_index=False).count()
Arms_Category=Arms_Category.sort_values(['id'], ascending=False)
Arms_Category
```

```python
pivot_Arms_Category = Arms_Category.pivot(index='arms_category',columns='manner_of_death')
pivot_Arms_Category.head()
```

```python
colors_list =['#5bc0de', '#d9534f']

ax = pivot_Arms_Category.plot(kind='bar', figsize=(15,4), width=0.8, color=colors_list, edgecolor=None)

for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy()
    ax.annotate(f'{height}', (x +width/2, y + height*1.02), ha='center')
```

determine death in relation to fleeing and arm category

```python
flee_arms=data[['id', 'arms_category', 'flee',]]
Flee_Arms= flee_arms.groupby(['flee', 'arms_category'],as_index=False).count()
Flee_Arms=Flee_Arms.sort_values(['flee'], ascending=False)
Flee_Arms
```

```python
pivot_Flee_Arms = Flee_Arms.pivot(index='arms_category',columns='flee')
pivot_Flee_Arms.head()
```

```python
plt.figure(figsize=(25,10))
sns.swarmplot(data=data, x="arms_category", y="id",hue="flee")
plt.xlabel("")
plt.ylabel("")
plt.title('Death based on Flee and Arms Catergory')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

```python
#analysis by states
plt.figure(figsize=(18,8))
plt.bar(data['state'].value_counts().index,data['state'].value_counts().values)
plt.title('Killings by based on states')
plt.ylabel('Number of Victims')
plt.xticks(fontsize=11)
plt.show()
```
